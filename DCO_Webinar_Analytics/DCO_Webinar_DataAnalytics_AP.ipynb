# Clustering (Unsupervised Learning)

- Grouping objects into classes so that similar objects are in the same class and dissimilar objects are in different class.
- Discover overall distribution patterns and relationships between attributes. 

# Calculate Gower Distance

```r
library(cluster)
library(factoextra)
library(NbClust)
set.seed(1234)
```

```r
gower_dist <- daisy(U_Data_Cleaned[,],
                    metric = "gower")
gower_mat <- as.matrix(gower_dist)
```

# Find the Optimum Number of clusters

```r
sil_width <- c()

for(i in 2:8){
  pam_fit <- pam(gower_dist,
                 diss = TRUE,
                 k = i)
  sil_width[i] <- pam_fit$silinfo$avg.width
}

plot(sil_width,xlab = "Number of clusters",
     ylab = "Silhouette Width")
lines(sil_width)
```

# Run clustering algorithm

```r
pam_fit <- pam(gower_dist, diss = T, k = 2)
clusplot(pam_fit)
```

# Association Rule Mining (Rule Based Learning)

- Initially developed for market basket analysis.
- Goal is to discover relationships between attributes.
- Uses include decision support, classification and clustering.

```r
U_MinLoc_Data
```

```r
U.associated.mins <- strsplit(as.character(U_MinLoc_Data$Query.Minerals),',')
U_associated_localities_minerals <- data.frame(locality=rep(U_MinLoc_Data$mindat.ID, sapply(U.associated.mins, FUN=length)), mineral=unlist(U.associated.mins),stringsAsFactors = F)

# remove leading & trailing spaces from mineral names
U_associated_localities_minerals$mineral <- sapply(U_associated_localities_minerals$mineral,function(x) {sub("^\\s+|^\\s+$","",x)})

U_associated.loc.min <- U_associated_localities_minerals
U_associated.loc.min$locality <- as.character(U_associated.loc.min$locality)
U_associated.loc.min.mat <- as.data.frame.matrix(table(U_associated.loc.min[,1:2]))

U_associated.loc.min.mat

U_associated_Min_Mat<-sapply(X = U_associated.loc.min.mat,FUN = function(x){replace(x, x > 0,1)})
#U_associated.loc.min.mat[which(U_associated.loc.min.mat != 0)] <- 1

U_associated_Min_Mat <- as.data.frame(U_associated_Min_Mat)
rownames(U_associated_Min_Mat) <- rownames(U_associated.loc.min.mat)

U_associated.Min.Mat<-as.data.frame(unclass(U_associated_Min_Mat))

#Logical Matrix Needed if only presence is to be considered. 
U_associated.Min.Mat<-sapply(U_associated.loc.min.mat, as.logical)
```

# Run Apriori Algorithm

```r
library(arules)
rules_support_0_008 <- apriori(data = U.associated.mins, parameter = list(supp = 0.008, conf = 0.4,target = "rules",maxtime = 0,maxlen = 8,minlen = 3))
#rules_support_0_0001 <- apriori(data = U_associated.Min.Mat, parameter = list(supp = 0.0001, conf = 0.7,target = "rules",maxtime = 0,maxlen = 30))

rules.sub.AP <- subset(rules_support_0_008, subset = ((lhs %pin% "Cervantite")))

inspect(tail(rules_support_0_008))
inspect(rules_support_0_008)
#Cervantite,Pyrite,Ferberite
```

##  4. Predictive Analytics

# Supervised Learning

- Training Data contains explicit examples of what the correct outputs should be for given inputs.


# Classification

- Identify to which of a set of classes a new observation belongs.
- Predicted values are commonly categorical, ordinal or discrete. 

# Model to predict redox state

```r
library(party)
m <- ctree(U_Data_Cleaned$U.redox ~ U_Data_Cleaned$Max.Age..Ma. + U_Data_Cleaned$Paragenetic.mode +  U_Data_Cleaned$v + U_Data_Cleaned$IG + U_Data_Cleaned$color + U_Data_Cleaned$luster + U_Data_Cleaned$hard + U_Data_Cleaned$crystal.system + U_Data_Cleaned$volume + U_Data_Cleaned$topology,data = U_Data_Cleaned)
plot(m)
```

# Prediction

```r
U_Redox_Pred<-predict(m,U_Data_Cleaned)

library(caret)
confusionMatrix(U_Redox_Pred,U_Data_Cleaned$U.redox)
```

# Regression

- An Approach to modelling the relationship between a dependent variable and one or more features.
- Predicted variable are commonly numeric and continous.

# Model to predict Maximum Age of the Mineral

```r
U_Data_Cleaned <- U_Data_Cleaned[!is.na(U_Data_Cleaned$Max.Age..Ma.),]
```

```r
m <- ctree(U_Data_Cleaned$Max.Age..Ma. ~ U_Data_Cleaned$U.redox + U_Data_Cleaned$Paragenetic.mode + U_Data_Cleaned$IG + U_Data_Cleaned$color + U_Data_Cleaned$luster + U_Data_Cleaned$hard + U_Data_Cleaned$crystal.system + U_Data_Cleaned$volume + U_Data_Cleaned$topology,data = U_Data_Cleaned)
plot(m)
```

```r
U_pred_Reg<-predict(m,U_Data_Cleaned)
plot(U_Data_Cleaned$Max.Age..Ma.-U_pred_Reg)
abline(h=0)
```
